```python
# For reproducibility
import random
import numpy as np
r_state = 42
random.seed(r_state) 
np.random.seed(r_state)

import geopandas as gpd
import pandas as pd
import numpy as np
from functools import reduce

import os

# Ruta donde están los CSV
ruta = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\Procesados\Distritos"
anio_2011 = "2011"  # 2011
anio_2021 = "2021"  # 2021

# Crear una lista para almacenar los DataFrames
dataframes_2011 = []
dataframes_2021 = []

# Iterar sobre los archivos en la ruta
for archivo in os.listdir(ruta):
    # Verificar si el archivo es un CSV y comienza con el año
    if archivo.endswith('.csv') and archivo.startswith(anio_2011):
        # Leer el archivo y agregarlo a la lista de DataFrames
        df = pd.read_csv(os.path.join(ruta, archivo),sep=";")
        dataframes_2011.append(df)
        print(df)
        
        
    if archivo.endswith('.csv') and archivo.startswith(anio_2021):
        # Leer el archivo y agregarlo a la lista de DataFrames
        df = pd.read_csv(os.path.join(ruta, archivo),sep=";")
        dataframes_2021.append(df)

# Unir todos los DataFrames por la columna 'NUMERO' utilizando reduce
df_2011 = reduce(lambda left, right: pd.merge(left, right, on="NUMERO"), dataframes_2011)
df_2021 = reduce(lambda left, right: pd.merge(left, right, on="NUMERO"), dataframes_2021)
#print(df_2011)
#print(df_2021)

#Incluir scores
path_scores = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Scores\Scores.csv"
df_scores = pd.read_csv(path_scores,sep=";")
df_scores_2011 =  df_scores[["NUMERO","SES_11","kw_pct_11","he_pct_11"]]
df_scores_2021 =  df_scores[["NUMERO","SES_21","kw_pct_21","he_pct_21"]]

df_2011 = pd.merge(df_2011, df_scores_2011, on='NUMERO')
df_2021 = pd.merge(df_2021, df_scores_2021, on='NUMERO')

print(df_2011)#10rows x 61
print(df_2021)#10rows x 61
```

```python
df_2011.to_csv(os.path.join(ruta, "df_2011.csv"),index=False,sep=";")
df_2021.to_csv(os.path.join(ruta, "df_2021.csv"),index=False,sep=";")
```
