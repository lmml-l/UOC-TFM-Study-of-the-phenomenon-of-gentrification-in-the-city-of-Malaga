```python
import geopandas as gpd
import pandas as pd
import numpy as np

## Configuraciones Iniciales
lugarDeNacimiento_seccion_2011_csv_path = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\06 - Lugar de Nacimiento\Secciones\Ocupados de 16 y más años_LugarDeNacimiento_Secciones_Censo_2011.csv"
lugarDeNacimiento_seccion_2021_csv_path = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\06 - Lugar de Nacimiento\Secciones\Ocupados de 16 y más años_LugarDeNacimiento_Secciones_Censo_2021.csv"
cartografia_seccion_shp_path = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Shapefiles\da_cartografiaSeccionCensal-25830\da_cartografiaSeccionCensalPolygon.shp"

## Lectura inicial de datos estado civil
df_lugarDeNacimiento_seccion_2011 =  pd.read_csv(lugarDeNacimiento_seccion_2011_csv_path,sep=";")
df_lugarDeNacimiento_seccion_2021 =  pd.read_csv(lugarDeNacimiento_seccion_2021_csv_path,sep=";")
df_cartografia_seccion = gpd.read_file(cartografia_seccion_shp_path)

# Ver las columnas
#print(df_lugarDeNacimiento_seccion_2011.columns)
#print(df_lugarDeNacimiento_seccion_2021.columns)

# Inspeccionar los primeros registros
#print(df_lugarDeNacimiento_seccion_2011.head())
#print(df_lugarDeNacimiento_seccion_2021.head())

#Cambio el nombre de la columna 'Sección ' a Seccion de residencia
df_lugarDeNacimiento_seccion_2021.rename(columns={'Sección ': 'Seccion de residencia'}, inplace=True)

#Add  America del norte y Oceania mismas columnas
df_lugarDeNacimiento_seccion_2011["América del Norte"] = 0
df_lugarDeNacimiento_seccion_2011["Oceanía"] = 0

#Se realiza una normalizacion de los csv de ocupacion en dos columnas estado civil y personas 
df_lugarDeNacimiento_seccion_2011_long = pd.melt(df_lugarDeNacimiento_seccion_2011, id_vars=["Municipio de residencia","Seccion de residencia"], var_name='Lugar de Nacimiento', value_name='Personas')
#print(df_lugarDeNacimiento_seccion_2011_long.columns)
#print(df_lugarDeNacimiento_seccion_2011_long.head())
df_lugarDeNacimiento_seccion_2021_long = pd.melt(df_lugarDeNacimiento_seccion_2021, id_vars=["Municipio de residencia","Seccion de residencia"], var_name='Lugar de Nacimiento', value_name='Personas')
#print(df_lugarDeNacimiento_seccion_2021_long.columns)
#print(df_lugarDeNacimiento_seccion_2021_long.head())

## Eliminar filas ##
#Eliminar filas de municipio Total
df_lugarDeNacimiento_seccion_2011_long = df_lugarDeNacimiento_seccion_2011_long[df_lugarDeNacimiento_seccion_2011_long['Municipio de residencia'] != 'Total']
df_lugarDeNacimiento_seccion_2021_long = df_lugarDeNacimiento_seccion_2021_long[df_lugarDeNacimiento_seccion_2021_long['Municipio de residencia'] != 'Total']
#Eliminar filas de Seccion de residencia Total
df_lugarDeNacimiento_seccion_2011_long = df_lugarDeNacimiento_seccion_2011_long[df_lugarDeNacimiento_seccion_2011_long['Seccion de residencia'] != 'Total']
df_lugarDeNacimiento_seccion_2021_long = df_lugarDeNacimiento_seccion_2021_long[df_lugarDeNacimiento_seccion_2021_long['Seccion de residencia'] != 'Total']
#Eliminar filas Nacimiento total
#print(df_lugarDeNacimiento_seccion_2011_long['Lugar de Nacimiento'].unique())
#print(df_lugarDeNacimiento_seccion_2021_long['Lugar de Nacimiento'].unique())
df_lugarDeNacimiento_seccion_2011_long = df_lugarDeNacimiento_seccion_2011_long[df_lugarDeNacimiento_seccion_2011_long['Lugar de Nacimiento'] != 'Total']
df_lugarDeNacimiento_seccion_2021_long = df_lugarDeNacimiento_seccion_2021_long[df_lugarDeNacimiento_seccion_2021_long['Lugar de Nacimiento'] != 'Total']
df_lugarDeNacimiento_seccion_2011_long = df_lugarDeNacimiento_seccion_2011_long[df_lugarDeNacimiento_seccion_2011_long['Lugar de Nacimiento'] != 'Apátrida']
df_lugarDeNacimiento_seccion_2021_long = df_lugarDeNacimiento_seccion_2021_long[df_lugarDeNacimiento_seccion_2021_long['Lugar de Nacimiento'] != 'Apátrida']


##df_lugarDeNacimiento_seccion_2011_long = df_lugarDeNacimiento_seccion_2011_long[df_lugarDeNacimiento_seccion_2011_long['Lugar de Nacimiento'] != 'América del Norte']
##df_lugarDeNacimiento_seccion_2021_long = df_lugarDeNacimiento_seccion_2021_long[df_lugarDeNacimiento_seccion_2021_long['Lugar de Nacimiento'] != 'América del Norte']

#Contamos el numero de valores que hay en estos casos
num_filas_en_blanco_2011 = (df_lugarDeNacimiento_seccion_2011_long['Personas'].isna()).sum()
num_filas_con_asterisco_2011 = (df_lugarDeNacimiento_seccion_2011_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2011 = (df_lugarDeNacimiento_seccion_2011_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2011)#0
#print(num_filas_con_asterisco_2011)#280
#print(num_filas_contiene_asterisco_2011)#250
num_filas_en_blanco_2021 = (df_lugarDeNacimiento_seccion_2021_long['Personas'].isna()).sum()
num_filas_con_asterisco_2021 = (df_lugarDeNacimiento_seccion_2021_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2021 = (df_lugarDeNacimiento_seccion_2021_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2021)#0
#print(num_filas_con_asterisco_2021)#0
#print(num_filas_contiene_asterisco_2021)#0

## Añadir columnas ##
#Añadir columna NUMSECCENS con el NUM de la seccion
df_lugarDeNacimiento_seccion_2011_long["NUMSECCENS"] = df_lugarDeNacimiento_seccion_2011_long["Seccion de residencia"].str[-5:].astype(float)
df_lugarDeNacimiento_seccion_2021_long["NUMSECCENS"] = df_lugarDeNacimiento_seccion_2021_long["Seccion de residencia"].str[-5:].astype(float)
#print(df_lugarDeNacimiento_seccion_2011_long.head())
#print(df_lugarDeNacimiento_seccion_2021_long.head())

## Gestion de Nulos y valores extraños ##
# Limpiar los datos: eliminar el asterisco al final, pero dejar intacto el '*'
df_lugarDeNacimiento_seccion_2011_long['Personas'] = df_lugarDeNacimiento_seccion_2011_long['Personas'].replace(r'(\d+)\*$', r'\1', regex=True)
df_lugarDeNacimiento_seccion_2021_long['Personas'] = df_lugarDeNacimiento_seccion_2021_long['Personas'].replace(r'(\d+)\*$', r'\1', regex=True)
#Sustituimos los valores * por NA
df_lugarDeNacimiento_seccion_2011_long['Personas'] = df_lugarDeNacimiento_seccion_2011_long['Personas'].replace('*', np.nan)
df_lugarDeNacimiento_seccion_2021_long['Personas'] = df_lugarDeNacimiento_seccion_2021_long['Personas'].replace('*', np.nan)
#Columna Personas como float
df_lugarDeNacimiento_seccion_2011_long['Personas'] = df_lugarDeNacimiento_seccion_2011_long['Personas'].astype(float)
df_lugarDeNacimiento_seccion_2021_long['Personas'] = df_lugarDeNacimiento_seccion_2021_long['Personas'].astype(float)

#Contamos el numero de valores que hay en estos casos
num_filas_en_blanco_2011 = (df_lugarDeNacimiento_seccion_2011_long['Personas'].isna()).sum()
num_filas_con_asterisco_2011 = (df_lugarDeNacimiento_seccion_2011_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2011 = (df_lugarDeNacimiento_seccion_2011_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2011)#244
#print(num_filas_con_asterisco_2011)#0
#print(num_filas_contiene_asterisco_2011)#0
num_filas_en_blanco_2021 = (df_lugarDeNacimiento_seccion_2021_long['Personas'].isna()).sum()
num_filas_con_asterisco_2021 = (df_lugarDeNacimiento_seccion_2021_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2021 = (df_lugarDeNacimiento_seccion_2021_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2021)#0
#print(num_filas_con_asterisco_2021)#0
#print(num_filas_contiene_asterisco_2021)#0

#Hay 244 valores vacios que sabemos eran * y segun el INE su valor esta entre 1-5. Usaremos el 3 como valor arbitrario (media)
# Calcular la desviación estándar antes de la imputación
desviacion_imputada_2011= df_lugarDeNacimiento_seccion_2011_long.groupby('Seccion de residencia')['Personas'].std()
#print("Desviación estándar por distrito antes de imputar 3:", desviacion_imputada_2011)
#Vamos a imputar un 3 y ver si hay mucha desviacion
df_lugarDeNacimiento_seccion_2011_long['Personas'] = df_lugarDeNacimiento_seccion_2011_long['Personas'].replace(np.nan, 3)
df_lugarDeNacimiento_seccion_2021_long['Personas'] = df_lugarDeNacimiento_seccion_2021_long['Personas'].replace(np.nan, 3)
# Calcular la desviación estándar despues de la imputación
desviacion_imputada_2011= df_lugarDeNacimiento_seccion_2011_long.groupby('Seccion de residencia')['Personas'].std()
#print("Desviación estándar por distrito despues de imputar 3:", desviacion_imputada_2011)

#En 2021 sumamos las categorias "Centro América y Caribe" y Sudamérica en una misma como en 2011 para que sean comparables
# Crear la categoria como 2011
def group(df):
    # Filtrar por los registros de divorciados y separados
    filtered = df[df["Lugar de Nacimiento"].isin(["Centro América y Caribe", "Sudamérica"])]
    
    # Agrupar y sumar las personas por Municipio y Seccion
    grouped = filtered.groupby(["Municipio de residencia", "Seccion de residencia","NUMSECCENS"], as_index=False)["Personas"].sum()
    grouped["Lugar de Nacimiento"] = "América Central, Sur y Caribe"

    return grouped

# Obtener los datos agrupados
grouped_data = group(df_lugarDeNacimiento_seccion_2021_long)

# Eliminar las filas originales y añadir la nueva categoría
df_lugarDeNacimiento_seccion_2021_long = pd.concat([
    df_lugarDeNacimiento_seccion_2021_long[~df_lugarDeNacimiento_seccion_2021_long["Lugar de Nacimiento"].isin(["Centro América y Caribe", "Sudamérica"])],
    grouped_data
], ignore_index=True)

#print(df_lugarDeNacimiento_seccion_2011_long.count())#324 registros
#print(df_lugarDeNacimiento_seccion_2021_long.count())#952 registros
#print(df_lugarDeNacimiento_seccion_2011_long.head())
#print(df_lugarDeNacimiento_seccion_2021_long.head())

##Calcular el total de personas por distrito y el porcentaje de Personas
df_lugarDeNacimiento_seccion_2011_long['Total_personas_distrito'] = df_lugarDeNacimiento_seccion_2011_long.groupby('Seccion de residencia')['Personas'].transform('sum')
df_lugarDeNacimiento_seccion_2021_long['Total_personas_distrito'] = df_lugarDeNacimiento_seccion_2021_long.groupby('Seccion de residencia')['Personas'].transform('sum')
#Calcular porcentaje sobre el total
df_lugarDeNacimiento_seccion_2011_long['Personas %'] = (df_lugarDeNacimiento_seccion_2011_long['Personas'] / df_lugarDeNacimiento_seccion_2011_long['Total_personas_distrito'])
df_lugarDeNacimiento_seccion_2021_long['Personas %'] = (df_lugarDeNacimiento_seccion_2021_long['Personas'] / df_lugarDeNacimiento_seccion_2021_long['Total_personas_distrito'])
#print(df_lugarDeNacimiento_seccion_2011_long[df_lugarDeNacimiento_seccion_2011_long['NUMSECCENS'] == 7092])
#print(df_lugarDeNacimiento_seccion_2011_long[df_lugarDeNacimiento_seccion_2011_long['NUMSECCENS'] == 3])
#print(df_lugarDeNacimiento_seccion_2011_long[df_lugarDeNacimiento_seccion_2011_long['NUMSECCENS'] == 5])
#print(df_lugarDeNacimiento_seccion_2021_long[df_lugarDeNacimiento_seccion_2021_long['NUMSECCENS'] == 7092])

## Unificar valores categorias Estudios en censos 2011 y 2021
mapping_dict_2011_2021 = {
    'Total': 'Total',
    'España': 'Española',
    'Otros Estados miembros de la Unión Europea': 'Unión Europea (sin España)',
    'Otros Estados de Europa fuera de la Unión Europea': 'Resto de Europa',
    'África': 'África',
    'América del Norte': 'América del Norte',
    'América Central, Sur y Caribe': 'América Central, Sur y Caribe',#Inicialmente en  2021 son dos categorias
    'Asia': 'Asia',
    'Oceanía': 'Oceanía'
}

df_lugarDeNacimiento_seccion_2011_long["Lugar de Nacimiento"] = df_lugarDeNacimiento_seccion_2011_long["Lugar de Nacimiento"].map(mapping_dict_2011_2021)
#print(df_lugarDeNacimiento_seccion_2011_long["Lugar de Nacimiento"].unique())
#print(df_lugarDeNacimiento_seccion_2021_long["Lugar de Nacimiento"].unique())

#Pivotar la tabla de nuevo para crear 8 variables, una por categoria de Lugar de nacimiento con el % ; una fila por distrito
df_lugarDeNacimiento_seccion_2011_long  = df_lugarDeNacimiento_seccion_2011_long.pivot(index='NUMSECCENS', columns='Lugar de Nacimiento', values='Personas %')
df_lugarDeNacimiento_seccion_2021_long  = df_lugarDeNacimiento_seccion_2021_long.pivot(index='NUMSECCENS', columns='Lugar de Nacimiento', values='Personas %')

print(df_lugarDeNacimiento_seccion_2011_long)
print(df_lugarDeNacimiento_seccion_2011_long.count())#54
print(df_lugarDeNacimiento_seccion_2021_long)
print(df_lugarDeNacimiento_seccion_2021_long.count())#136
```
