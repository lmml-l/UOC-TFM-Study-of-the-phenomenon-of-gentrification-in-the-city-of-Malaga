```python
import geopandas as gpd
import pandas as pd
import numpy as np

## Configuraciones Iniciales
estadoCivil_seccion_2011_csv_path = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\05 - Estado Civil\Secciones\Ocupados de 16 y más años_EstadoCivil_Secciones_Censo_2011.csv"
estadoCivil_seccion_2021_csv_path = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\05 - Estado Civil\Secciones\Ocupados de 16 y más años_EstadoCivil_Secciones_Censo_2021.csv"
cartografia_seccion_shp_path = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Shapefiles\da_cartografiaSeccionCensal-25830\da_cartografiaSeccionCensalPolygon.shp"

## Lectura inicial de datos estado civil
df_estadoCivil_seccion_2011 =  pd.read_csv(estadoCivil_seccion_2011_csv_path,sep=";")
df_estadoCivil_seccion_2021 =  pd.read_csv(estadoCivil_seccion_2021_csv_path,sep=";")
df_cartografia_seccion = gpd.read_file(cartografia_seccion_shp_path)

# Ver las columnas
#print(df_estadoCivil_seccion_2011.columns)
#print(df_estadoCivil_seccion_2021.columns)

# Inspeccionar los primeros registros
#print(df_estadoCivil_seccion_2011.head())
#print(df_estadoCivil_seccion_2021.head())

#Cambio el nombre de la columna 'Sección ' a Seccion de residencia
df_estadoCivil_seccion_2021.rename(columns={'Sección ': 'Seccion de residencia'}, inplace=True)

#Se realiza una normalizacion de los csv de ocupacion en dos columnas estado civil y personas 
df_estadoCivil_seccion_2011_long = pd.melt(df_estadoCivil_seccion_2011, id_vars=["Municipio de residencia","Seccion de residencia"], var_name='Estado Civil', value_name='Personas')
#print(df_estadoCivil_seccion_2011_long.columns)
#print(df_estadoCivil_seccion_2011_long.head())
df_estadoCivil_seccion_2021_long = pd.melt(df_estadoCivil_seccion_2021, id_vars=["Municipio de residencia","Seccion de residencia"], var_name='Estado Civil', value_name='Personas')
#print(df_estadoCivil_seccion_2021_long.columns)
#print(df_estadoCivil_seccion_2021_long.head())

## Eliminar filas ##
#Eliminar filas de municipio Total
df_estadoCivil_seccion_2011_long = df_estadoCivil_seccion_2011_long[df_estadoCivil_seccion_2011_long['Municipio de residencia'] != 'Total']
df_estadoCivil_seccion_2021_long = df_estadoCivil_seccion_2021_long[df_estadoCivil_seccion_2021_long['Municipio de residencia'] != 'Total']
#Eliminamos las filas de residencia Total
df_estadoCivil_seccion_2011_long = df_estadoCivil_seccion_2011_long [df_estadoCivil_seccion_2011_long ['Seccion de residencia'] != 'Total']
df_estadoCivil_seccion_2021_long = df_estadoCivil_seccion_2021_long [df_estadoCivil_seccion_2021_long ['Seccion de residencia'] != 'Total']
#Eliminamos las filas de estudio Total 
df_estadoCivil_seccion_2011_long = df_estadoCivil_seccion_2011_long [df_estadoCivil_seccion_2011_long ['Estado Civil'] != 'Total']
df_estadoCivil_seccion_2021_long = df_estadoCivil_seccion_2021_long [df_estadoCivil_seccion_2021_long ['Estado Civil'] != 'Total']
df_estadoCivil_seccion_2011_long = df_estadoCivil_seccion_2011_long [df_estadoCivil_seccion_2011_long ['Estado Civil'] != 'No consta']
df_estadoCivil_seccion_2021_long = df_estadoCivil_seccion_2021_long [df_estadoCivil_seccion_2021_long ['Estado Civil'] != 'No consta']
#print(df_estadoCivil_seccion_2011_long.head())
#print(df_estadoCivil_seccion_2021_long.head())

#Contamos el numero de valores que hay en estos casos
num_filas_en_blanco_2011 = (df_estadoCivil_seccion_2011_long['Personas'].isna()).sum()
num_filas_con_asterisco_2011 = (df_estadoCivil_seccion_2011_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2011 = (df_estadoCivil_seccion_2011_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2011)#0
#print(num_filas_con_asterisco_2011)#0
#print(num_filas_contiene_asterisco_2011)#80
num_filas_en_blanco_2021 = (df_estadoCivil_seccion_2021_long['Personas'].isna()).sum()
num_filas_con_asterisco_2021 = (df_estadoCivil_seccion_2021_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2021 = (df_estadoCivil_seccion_2021_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2021)#0
#print(num_filas_con_asterisco_2021)#0
#print(num_filas_contiene_asterisco_2021)#0

## Añadir columnas ##
#Añadir columna NUMSECCENS con el NUM de la seccion
df_estadoCivil_seccion_2011_long["NUMSECCENS"] = df_estadoCivil_seccion_2011_long["Seccion de residencia"].str[-5:].astype(float)
df_estadoCivil_seccion_2021_long["NUMSECCENS"] = df_estadoCivil_seccion_2021_long["Seccion de residencia"].str[-5:].astype(float)

# Limpiar los datos: eliminar el asterisco al final, pero dejar intacto el '*'
df_estadoCivil_seccion_2011_long['Personas'] = df_estadoCivil_seccion_2011_long['Personas'].replace(r'(\d+)\*$', r'\1', regex=True)
df_estadoCivil_seccion_2021_long['Personas'] = df_estadoCivil_seccion_2021_long['Personas'].replace(r'(\d+)\*$', r'\1', regex=True)
#Columna Personas como float
df_estadoCivil_seccion_2011_long['Personas'] = df_estadoCivil_seccion_2011_long['Personas'].astype(float)
df_estadoCivil_seccion_2021_long['Personas'] = df_estadoCivil_seccion_2021_long['Personas'].astype(float)

#Contamos el numero de valores que hay en estos casos
num_filas_en_blanco_2011 = (df_estadoCivil_seccion_2011_long['Personas'].isna()).sum()
num_filas_con_asterisco_2011 = (df_estadoCivil_seccion_2011_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2011 = (df_estadoCivil_seccion_2011_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2011)#0
#print(num_filas_con_asterisco_2011)#0
#print(num_filas_contiene_asterisco_2011)#0
num_filas_en_blanco_2021 = (df_estadoCivil_seccion_2021_long['Personas'].isna()).sum()
num_filas_con_asterisco_2021 = (df_estadoCivil_seccion_2021_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2021 = (df_estadoCivil_seccion_2021_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2021)#0
#print(num_filas_con_asterisco_2021)#0
#print(num_filas_contiene_asterisco_2021)#0

## Unificar valores categorias Estado Civil en censos 2011 y 2021
#Los valores que toma el campo Estado Civil son distintos
#print(df_estadoCivil_seccion_2011_long['Estado Civil'].unique())#Tamaño 5 ['Soltero/a' 'Casado/a' 'Separado/a' 'Divorciado/a' 'Viudo/a']
#print(df_estadoCivil_seccion_2021_long['Estado Civil'].unique())#Tamaño 4 ['Soltero/a' 'Casado/a' 'Viudo/a' 'Divorciado/a o Separado/a']

#En 2011 sumamos las categorias Separado y Divorciado en una misma como en 2021 para que sean comparables
# Crear una nueva categoría "divorciados y separados"
def group_divorced_separated(df):
    # Filtrar por los registros de divorciados y separados
    filtered = df[df["Estado Civil"].isin(["Separado/a", "Divorciado/a"])]
    
    # Agrupar y sumar las personas por Municipio y Distrito
    grouped = filtered.groupby(["Municipio de residencia", "Seccion de residencia","NUMSECCENS"], as_index=False)["Personas"].sum()
    grouped["Estado Civil"] = "Divorciado/a o Separado/a"

    return grouped

# Obtener los datos agrupados
grouped_data = group_divorced_separated(df_estadoCivil_seccion_2011_long)

# Eliminar las filas originales de divorciados y separados y añadir la nueva categoría
df_estadoCivil_seccion_2011_long = pd.concat([
    df_estadoCivil_seccion_2011_long[~df_estadoCivil_seccion_2011_long["Estado Civil"].isin(["Separado/a", "Divorciado/a"])],
    grouped_data
], ignore_index=True)

## Pasar los valores a relativo (Porcentajes) ##
#Calcular el total de personas por distrito
df_estadoCivil_seccion_2011_long['Total_personas_distrito'] = df_estadoCivil_seccion_2011_long.groupby('Seccion de residencia')['Personas'].transform('sum')
df_estadoCivil_seccion_2021_long['Total_personas_distrito'] = df_estadoCivil_seccion_2021_long.groupby('Seccion de residencia')['Personas'].transform('sum')
#Calcular porcentaje sobre el total
df_estadoCivil_seccion_2011_long['Personas %'] = (df_estadoCivil_seccion_2011_long['Personas'] / df_estadoCivil_seccion_2011_long['Total_personas_distrito'])
df_estadoCivil_seccion_2021_long['Personas %'] = (df_estadoCivil_seccion_2021_long['Personas'] / df_estadoCivil_seccion_2021_long['Total_personas_distrito'])

#print(df_estadoCivil_seccion_2011_long[df_estadoCivil_seccion_2011_long["NUMSECCENS"] == 2003])
#print(df_estadoCivil_seccion_2021_long[df_estadoCivil_seccion_2021_long["NUMSECCENS"] == 2003])

#Pivotar la tabla de nuevo para crear 4 variables, una por categoria de Estado Civil con el % ; una fila por seccion
df_estadoCivil_seccion_2011_long  = df_estadoCivil_seccion_2011_long.pivot(index='NUMSECCENS', columns='Estado Civil', values='Personas %')
df_estadoCivil_seccion_2021_long  = df_estadoCivil_seccion_2021_long.pivot(index='NUMSECCENS', columns='Estado Civil', values='Personas %')

print(df_estadoCivil_seccion_2011_long)
print(df_estadoCivil_seccion_2011_long.count())#54
print(df_estadoCivil_seccion_2021_long)
print(df_estadoCivil_seccion_2021_long.count())#136
```
