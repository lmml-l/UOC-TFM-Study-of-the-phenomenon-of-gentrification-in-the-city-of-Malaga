```python
import geopandas as gpd
import pandas as pd
import numpy as np

## Configuraciones Iniciales
estudios_detalle_seccion_2011_csv_path = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\04  - Estudios\Secciones\Ocupados de 16 y más años_Estudios_Detalle_Secciones_Censo_2011.csv"
estudios_detalle_seccion_2021_csv_path = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\04  - Estudios\Secciones\Ocupados de 16 y más años_Estudios_Detalle_Secciones_Censo_2021.csv"
cartografia_seccion_shp_path = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Shapefiles\da_cartografiaSeccionCensal-25830\da_cartografiaSeccionCensalPolygon.shp"

## Lectura inicial de datos estudios
df_estudios_detalle_seccion_2011 =  pd.read_csv(estudios_detalle_seccion_2011_csv_path,sep=";")
df_estudios_detalle_seccion_2021 =  pd.read_csv(estudios_detalle_seccion_2021_csv_path,sep=";")
df_cartografia_seccion = gpd.read_file(cartografia_seccion_shp_path)

# Ver las columnas
#print(df_estudios_detalle_seccion_2011.columns)
#print(df_estudios_detalle_seccion_2021.columns)
#print(df_cartografia_seccion.columns)

# Inspeccionar los primeros registros
#print(df_estudios_detalle_seccion_2011.head())
#print(df_estudios_detalle_seccion_2021.head())
print(df_cartografia_seccion.sort_values(by="NUMSECCENS").head())
print(df_cartografia_seccion.sort_values(by="NUMSECCENS").tail())

df_estudios_detalle_seccion_2011_long = pd.melt(df_estudios_detalle_seccion_2011, id_vars=["Municipio de residencia","Seccion de residencia"], var_name='Estudios', value_name='Personas')
#print(df_estudios_detalle_seccion_2011_long.columns)
#print(df_estudios_detalle_seccion_2011_long.head())
df_estudios_detalle_seccion_2021_long = pd.melt(df_estudios_detalle_seccion_2021, id_vars=["Municipio de residencia","Sección "], var_name='Estudios', value_name='Personas')
#print(df_estudios_detalle_seccion_2021_long.columns)
#print(df_estudios_detalle_seccion_2021_long.head())

#Cambio el nombre de la columna 'Sección ' a Seccion de residencia
df_estudios_detalle_seccion_2021_long.rename(columns={'Sección ': 'Seccion de residencia'}, inplace=True)

## Eliminar filas ##
#Eliminar filas de municipio Total
df_estudios_detalle_seccion_2011_long = df_estudios_detalle_seccion_2011_long[df_estudios_detalle_seccion_2011_long['Municipio de residencia'] != 'Total']
df_estudios_detalle_seccion_2021_long = df_estudios_detalle_seccion_2021_long[df_estudios_detalle_seccion_2021_long['Municipio de residencia'] != 'Total']
#Eliminamos las filas de residencia Total
df_estudios_detalle_seccion_2011_long = df_estudios_detalle_seccion_2011_long [df_estudios_detalle_seccion_2011_long ['Seccion de residencia'] != 'Total']
df_estudios_detalle_seccion_2021_long = df_estudios_detalle_seccion_2021_long [df_estudios_detalle_seccion_2021_long ['Seccion de residencia'] != 'Total']
#Eliminamos las filas de estudio Total 
df_estudios_detalle_seccion_2011_long = df_estudios_detalle_seccion_2011_long [df_estudios_detalle_seccion_2011_long ['Estudios'] != 'Total']
df_estudios_detalle_seccion_2021_long = df_estudios_detalle_seccion_2021_long [df_estudios_detalle_seccion_2021_long ['Estudios'] != 'Total']
#print(df_estudios_detalle_seccion_2011_long.head())
#print(df_estudios_detalle_seccion_2021_long.head())

#Contamos el numero de valores que hay en estos casos
num_filas_en_blanco_2011 = (df_estudios_detalle_seccion_2011_long['Personas'].isna()).sum()
num_filas_con_asterisco_2011 = (df_estudios_detalle_seccion_2011_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2011 = (df_estudios_detalle_seccion_2011_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2011)#0
#print(num_filas_con_asterisco_2011)#4209
#print(num_filas_contiene_asterisco_2011)#4209
num_filas_en_blanco_2021 = (df_estudios_detalle_seccion_2021_long['Personas'].isna()).sum()
num_filas_con_asterisco_2021 = (df_estudios_detalle_seccion_2021_long['Personas'] == "*").sum()
num_filas_contiene_asterisco_2021 = (df_estudios_detalle_seccion_2021_long['Personas'].apply(str).str.contains(r'\*')).sum()
#print(num_filas_en_blanco_2021)#0
#print(num_filas_con_asterisco_2021)#0
#print(num_filas_contiene_asterisco_2021)#0

## Añadir columnas ##
#Añadir columna NUMSECCENS con el NUM de la seccion
df_estudios_detalle_seccion_2011_long["NUMSECCENS"] = df_estudios_detalle_seccion_2011_long["Seccion de residencia"].str[-5:].astype(float)
df_estudios_detalle_seccion_2021_long["NUMSECCENS"] = df_estudios_detalle_seccion_2021_long["Seccion de residencia"].str[-5:].astype(float)

## Gestion de Nulos y valores extraños ##
#Sustituimos los valores * por NA
df_estudios_detalle_seccion_2011_long['Personas'] = df_estudios_detalle_seccion_2011_long['Personas'].replace('*', np.nan)
df_estudios_detalle_seccion_2021_long['Personas'] = df_estudios_detalle_seccion_2021_long['Personas'].replace('*', np.nan)
#Columna Personas como float
df_estudios_detalle_seccion_2011_long['Personas'] = df_estudios_detalle_seccion_2011_long['Personas'].astype(float)
df_estudios_detalle_seccion_2021_long['Personas'] = df_estudios_detalle_seccion_2021_long['Personas'].astype(float)
#Contamos el numero de valores tras los nuevos ajustes
num_filas_en_blanco_2011 = (df_estudios_detalle_seccion_2011_long['Personas'].isna()).sum()
num_filas_con_asterisco_2011 = (df_estudios_detalle_seccion_2011_long['Personas'] == "*").sum()
#print(num_filas_en_blanco_2011)#4209
#print(num_filas_con_asterisco_2011)#0

## Calcular % Personas estudios por seccion con valor medio 3 ##
#Los valores con * tienen un valor entre 1-5 según el INE. Utilizaremos la media (3) como estimacion
# Calcular la desviación estándar antes de la imputación
desviacion_imputada_2011= df_estudios_detalle_seccion_2011_long.groupby('Seccion de residencia')['Personas'].std()
#print("Desviación estándar por distrito antes de imputar 3:", desviacion_imputada_2011)
#No obstante vamos a ver si la desviacion del dato cambia mucho respecto a lo anterior usando el 3
df_estudios_detalle_seccion_2011_long['Personas'] = df_estudios_detalle_seccion_2011_long['Personas'].replace(np.nan, 3)
df_estudios_detalle_seccion_2021_long['Personas'] = df_estudios_detalle_seccion_2021_long['Personas'].replace(np.nan, 3)
desviacion_imputada_2011= df_estudios_detalle_seccion_2011_long.groupby('Seccion de residencia')['Personas'].std()
#print("Desviación estándar por distrito despues de imputar 3:", desviacion_imputada_2011)

#Calcular el total de personas por seccion
df_estudios_detalle_seccion_2011_long['Total_personas_seccion'] = df_estudios_detalle_seccion_2011_long.groupby('Seccion de residencia')['Personas'].transform('sum')
df_estudios_detalle_seccion_2021_long['Total_personas_seccion'] = df_estudios_detalle_seccion_2021_long.groupby('Seccion de residencia')['Personas'].transform('sum')
#Calcular porcentaje sobre el total
df_estudios_detalle_seccion_2011_long['Personas %'] = (df_estudios_detalle_seccion_2011_long['Personas'] / df_estudios_detalle_seccion_2011_long['Total_personas_seccion'])
df_estudios_detalle_seccion_2021_long['Personas %'] = (df_estudios_detalle_seccion_2021_long['Personas'] / df_estudios_detalle_seccion_2021_long['Total_personas_seccion'])
#print(df_estudios_detalle_seccion_2011_long[df_estudios_detalle_seccion_2011_long['NUMSECCENS'] == 1001])
#print(df_estudios_detalle_seccion_2011_long[df_estudios_detalle_seccion_2011_long['NUMSECCENS'] == 3001])
#print(df_estudios_detalle_seccion_2011_long[df_estudios_detalle_seccion_2011_long['NUMSECCENS'] == 5001])
#print(df_estudios_detalle_seccion_2021_long[df_estudios_detalle_seccion_2021_long['NUMSECCENS'] == 1001])

# Diccionario de mapeo 2011 a 2021
mapping_dict_2011_2021 = {
    'Total': 'Total',
    'Analfabetos': 'Analfabetos',
    'Sin estudios': 'Estudios primarios incompletos',
    'Fue a la escuela 5 años o más pero no llegó al último curso de ESO, EGB o Bachiller Elemental': 'Educación primaria',
    'Llegó al último curso de ESO, EGB o Bachiller Elemental o tiene el Certificado de Escolaridad o de Estudios Primarios': 'Primera etapa de educación secundaria y similar',
    'Bachiller, BUP, Bachiller Superior, COU, PREU': 'Segunda etapa de educación secundaria con orientación general',
    'FP grado medio, FP I, Oficialia Industrial o equivalente, Grado Medio de Música y Danza, Certificados de Escuelas Oficiales de Idiomas': 'Segunda etapa de educación secundaria con orientación profesional',
    'FP grado superior, FPII, Maestria industrial o equivalente': 'Educación postsecundaria no superior',
    'Diplomatura universitaria, Aquitectura Técnica, Ingeniería Técnica o equivalente': 'Enseñanzas de formación profesional, artes plásticas y diseño y deportivas de grado superior y equivalentes',
    'Grado Universitario o equivalente': 'Grados universitarios de 240 créditos ECTS, diplomados universitarios, títulos propios universitarios de experto o especialista',
    'Licenciatura, Arquitectura, Ingeniería o equivalente': 'Grados universitarios de más de 240 créditos ECTS, licenciados',
    'Máster oficial univeritario (a partir de 2006) , Especialidades Médicas o análogas': 'Másteres, especialidades en Ciencias de la Salud por el sistema de residencia y similares',
    'Doctorado': 'Doctorado universitario'
}

#Mapear mismas columnas entre censos
df_estudios_detalle_seccion_2011_long['Estudios'] = df_estudios_detalle_seccion_2011_long['Estudios'].map(mapping_dict_2011_2021)
#print(df_estudios_detalle_seccion_2011_long['Estudios'].unique())
#print(df_estudios_detalle_seccion_2021_long['Estudios'].unique())

#Crear nueva columna Id_Estudios
mapping_dict_estudios_id_estudios = {
    'Total': -1,
    'Analfabetos': 1,
    'Estudios primarios incompletos': 2,
    'Educación primaria': 3,
    'Primera etapa de educación secundaria y similar': 4,
    'Segunda etapa de educación secundaria con orientación general': 5,
    'Segunda etapa de educación secundaria con orientación profesional': 6,
    'Educación postsecundaria no superior': 7,
    'Enseñanzas de formación profesional, artes plásticas y diseño y deportivas de grado superior y equivalentes': 8,
    'Grados universitarios de 240 créditos ECTS, diplomados universitarios, títulos propios universitarios de experto o especialista': 9,
    'Grados universitarios de más de 240 créditos ECTS, licenciados' : 10,
    'Másteres, especialidades en Ciencias de la Salud por el sistema de residencia y similares':  11,
    'Doctorado universitario': 12
}

df_estudios_detalle_seccion_2011_long['Id_Estudios'] = df_estudios_detalle_seccion_2011_long['Estudios'].map(mapping_dict_estudios_id_estudios).astype(float)
df_estudios_detalle_seccion_2021_long['Id_Estudios'] = df_estudios_detalle_seccion_2021_long['Estudios'].map(mapping_dict_estudios_id_estudios).astype(float)

print(df_estudios_detalle_seccion_2011_long.head())
print(df_estudios_detalle_seccion_2021_long.head())
```

```python
#Export estudios top
#Los estudios top estan entre el id 9 al 12 (equivalente a NVQ4 o superior)
df_estudios_detalle_seccion_2011_long_high_educated = df_estudios_detalle_seccion_2011_long[df_estudios_detalle_seccion_2011_long['Id_Estudios'].isin([ 9, 10, 11, 12])]
df_estudios_detalle_seccion_2021_long_high_educated = df_estudios_detalle_seccion_2021_long[df_estudios_detalle_seccion_2021_long['Id_Estudios'].isin([ 9, 10, 11, 12])]

# Agrupar por distrito y sumar 'Personas' y 'Porcentaje'
df_agrupado_2011 = df_estudios_detalle_seccion_2011_long_high_educated.groupby(['NUMSECCENS','Seccion de residencia'], as_index=False).agg({
    'Personas %': 'sum'
})

df_agrupado_2021 = df_estudios_detalle_seccion_2021_long_high_educated.groupby(['NUMSECCENS','Seccion de residencia'], as_index=False).agg({
    'Personas %': 'sum'
})

#Renombrar columna
df_agrupado_2011.rename(columns={'Personas %': 'he_pct'}, inplace=True)
df_agrupado_2021.rename(columns={'Personas %': 'he_pct'}, inplace=True)

print(df_agrupado_2011)
print(df_agrupado_2021)

df_agrupado_2011.to_csv(r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\04  - Estudios\Secciones\2011_Secciones_Estudios_Detalle_V2.csv",index=False,sep=";")
df_agrupado_2021.to_csv(r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\04  - Estudios\Secciones\2021_Secciones_Estudios_Detalle_V2.csv",index=False,sep=";")
```
