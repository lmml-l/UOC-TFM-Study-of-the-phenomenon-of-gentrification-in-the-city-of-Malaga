```python
# For reproducibility
import random
import numpy as np
r_state = 42
random.seed(r_state) 
np.random.seed(r_state)

import geopandas as gpd
import pandas as pd
import numpy as np
from functools import reduce

import os
# Ruta donde están los CSV
ruta = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Variables\Procesados\Secciones"
anio_2011 = "2011"  # 2011
anio_2021 = "2021"  # 2021

# Crear una lista para almacenar los DataFrames
dataframes_2011 = []
dataframes_2021 = []

# Iterar sobre los archivos en la ruta
for archivo in os.listdir(ruta):
    # Verificar si el archivo es un CSV y comienza con el año
    if archivo.endswith('.csv') and archivo.startswith(anio_2011):
        # Leer el archivo y agregarlo a la lista de DataFrames
        df = pd.read_csv(os.path.join(ruta, archivo),sep=";")
        dataframes_2011.append(df)
        
        
    if archivo.endswith('.csv') and archivo.startswith(anio_2021):
        # Leer el archivo y agregarlo a la lista de DataFrames
        df = pd.read_csv(os.path.join(ruta, archivo),sep=";")
        dataframes_2021.append(df)

# Unir todos los DataFrames por la columna 'NUMSECCENS' utilizando reduce 
df_2011 = reduce(lambda left, right: pd.merge(left, right, on="NUMSECCENS"), dataframes_2011)
df_2021 = reduce(lambda left, right: pd.merge(left, right, on="NUMSECCENS"), dataframes_2021)
#print(df_2011)#35rows x 56 columns Tenemos 35 secciones con todos los datos
#print(df_2021)#123rows x 56 columns Tenemos 123 secciones con todos los datos hasta aqui

#Incluir scores
path_scores = r"C:\Users\Usuario\OneDrive\Escritorio\UOC\TFM\PEC3  - Implementacion\WorkStation\Datos\Scores\Scores-Secciones.csv"
df_scores = pd.read_csv(path_scores,sep=";")
df_scores_2011 =  df_scores[["NUMSECCENS","SES_11","kw_pct_11","he_pct_11","median_price_inf_11"]]
df_scores_2021 =  df_scores[["NUMSECCENS","SES_21","kw_pct_21","he_pct_21","median_price_inf_21"]]

df_2011 = pd.merge(df_2011, df_scores_2011, on='NUMSECCENS')
df_2021 = pd.merge(df_2021, df_scores_2021, on='NUMSECCENS')

print(df_2011)#35rows x 60
print(df_2021)#115rows x 60
```

```python
df_2011.to_csv(os.path.join(ruta, "df_2011.csv"),index=False,sep=";")
df_2021.to_csv(os.path.join(ruta, "df_2021.csv"),index=False,sep=";")
```

```python
# Unir todos los DataFrames por la columna 'NUMSECCENS' utilizando reduce left
df_2011_left = reduce(lambda left, right: pd.merge(left, right, on="NUMSECCENS", how="left"), dataframes_2011)
df_2021_left = reduce(lambda left, right: pd.merge(left, right, on="NUMSECCENS", how="left"), dataframes_2021)

df_2011_left = pd.merge(df_2011_left, df_scores_2011, on='NUMSECCENS')
df_2021_left = pd.merge(df_2021_left, df_scores_2021, on='NUMSECCENS')

print(df_2011)#35rows x 60
print(df_2021)#115rows x 60
```
